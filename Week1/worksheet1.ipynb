```python
#Problem - 3: Array Operations
import numpy as np

# Initialize the arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

print("Initial Arrays:")
print(f"x = \n{x}")
print(f"y = \n{y}")
print(f"v = {v}")
print(f"w = {w}")
print(f"x shape: {x.shape}, y shape: {y.shape}")
print(f"v shape: {v.shape}, w shape: {w.shape}")

# 1. Add the two arrays
print("\n1. Array Addition:")
print("x + y =")
print(x + y)

# 2. Subtract the two arrays
print("\n2. Array Subtraction:")
print("x - y =")
print(x - y)
print("y - x =")
print(y - x)

# 3. Multiply the array with any integers
print("\n3. Array Multiplication with scalars:")
print("2 * x =")
print(2 * x)
print("3 * y =")
print(3 * y)
print("0.5 * v =", 0.5 * v)

# 4. Find the square of each element of the array
print("\n4. Element-wise Square:")
print("x² =")
print(x ** 2)
print("y² =")
print(y ** 2)
print("v² =", v ** 2)

# Alternative method using np.square()
print("\nUsing np.square():")
print("np.square(x) =")
print(np.square(x))

# 5. Find the dot product
print("\n5. Dot Products:")

# Dot product between v and w
dot_vw = np.dot(v, w)
print(f"v · w = {dot_vw}")
print(f"Calculation: ({v[0]}×{w[0]}) + ({v[1]}×{w[1]}) = {v[0]*w[0]} + {v[1]*w[1]} = {dot_vw}")

# Dot product between x and v
dot_xv = np.dot(x, v)
print(f"\nx · v = {dot_xv}")
print("Calculation:")
print(f"Row 1: ({x[0,0]}×{v[0]}) + ({x[0,1]}×{v[1]}) = {x[0,0]*v[0]} + {x[0,1]*v[1]} = {x[0,0]*v[0] + x[0,1]*v[1]}")
print(f"Row 2: ({x[1,0]}×{v[0]}) + ({x[1,1]}×{v[1]}) = {x[1,0]*v[0]} + {x[1,1]*v[1]} = {x[1,0]*v[0] + x[1,1]*v[1]}")

# Dot product between x and y
dot_xy = np.dot(x, y)
print(f"\nx · y = \n{dot_xy}")
print("Calculation:")
print(f"Element [0,0]: ({x[0,0]}×{y[0,0]}) + ({x[0,1]}×{y[1,0]}) = {x[0,0]*y[0,0]} + {x[0,1]*y[1,0]} = {x[0,0]*y[0,0] + x[0,1]*y[1,0]}")
print(f"Element [0,1]: ({x[0,0]}×{y[0,1]}) + ({x[0,1]}×{y[1,1]}) = {x[0,0]*y[0,1]} + {x[0,1]*y[1,1]} = {x[0,0]*y[0,1] + x[0,1]*y[1,1]}")
print(f"Element [1,0]: ({x[1,0]}×{y[0,0]}) + ({x[1,1]}×{y[1,0]}) = {x[1,0]*y[0,0]} + {x[1,1]*y[1,0]} = {x[1,0]*y[0,0] + x[1,1]*y[1,0]}")
print(f"Element [1,1]: ({x[1,0]}×{y[0,1]}) + ({x[1,1]}×{y[1,1]}) = {x[1,0]*y[0,1]} + {x[1,1]}×{y[1,1]} = {x[1,0]*y[0,1] + x[1,1]*y[1,1]}")

# 6. Concatenation
print("\n6. Array Concatenation:")

# Concatenate x and y along row (axis=0)
concat_row = np.concatenate((x, y), axis=0)
print("Concatenate x and y along rows (axis=0):")
print(concat_row)

# Alternative using vstack
concat_vstack = np.vstack((x, y))
print("\nUsing np.vstack((x, y)):")
print(concat_vstack)

# Concatenate v and w along column (axis=1)
# First we need to make them 2D arrays for column concatenation
v_2d = v.reshape(-1, 1)  # Convert to column vector
w_2d = w.reshape(-1, 1)  # Convert to column vector
concat_col = np.concatenate((v_2d, w_2d), axis=1)
print("\nConcatenate v and w along columns (axis=1):")
print(concat_col)

# Alternative using hstack
concat_hstack = np.hstack((v.reshape(-1, 1), w.reshape(-1, 1)))
print("\nUsing np.hstack for v and w:")
print(concat_hstack)

# 7. Concatenate x and v
print("\n7. Concatenate x and v:")
print(f"x shape: {x.shape}")
print(f"v shape: {v.shape}")

try:
    result = np.concatenate((x, v))
    print("Success:", result)
except ValueError as e:
    print(f"Error: {e}")
    print("Explanation: Cannot concatenate arrays with different dimensions.")
    print("x is 2D (2x2) while v is 1D (2,). They must have the same number of dimensions.")

```

```python
#Problem - 4: Matrix Operations
import numpy as np

# Initialize matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Initial Matrices:")
print(f"A = \n{A}")
print(f"B = \n{B}")

# 1. Prove A.A⁻¹ = I
print("\n1. Prove A.A⁻¹ = I:")
print("=" * 50)

try:
    # Calculate inverse of A
    A_inv = np.linalg.inv(A)
    print(f"A⁻¹ = \n{A_inv}")
    
    # Multiply A with its inverse
    identity_check = np.dot(A, A_inv)
    print(f"\nA × A⁻¹ = \n{identity_check}")
    
    # Create actual identity matrix for comparison
    I = np.eye(2)
    print(f"\nActual Identity matrix I = \n{I}")
    
    # Check if they are close (due to floating point precision)
    is_identity = np.allclose(identity_check, I)
    print(f"\nIs A × A⁻¹ equal to I? {is_identity}")
    print(f"Maximum difference: {np.max(np.abs(identity_check - I)):.10f}")
    
except np.linalg.LinAlgError:
    print("Matrix A is singular, cannot compute inverse")

# 2. Prove AB ≠ BA
print("\n2. Prove AB ≠ BA:")
print("=" * 50)

AB = np.dot(A, B)
BA = np.dot(B, A)

print(f"A × B = \n{AB}")
print(f"B × A = \n{BA}")

are_equal = np.array_equal(AB, BA)
print(f"\nIs AB equal to BA? {are_equal}")

if not are_equal:
    print("✓ Proved: Matrix multiplication is not commutative (AB ≠ BA)")

# 3. Prove (AB)ᵀ = BᵀAᵀ
print("\n3. Prove (AB)ᵀ = BᵀAᵀ:")
print("=" * 50)

AB_transpose = AB.T
print(f"(AB)ᵀ = \n{AB_transpose}")

B_transpose = B.T
A_transpose = A.T
BTA_transpose = np.dot(B_transpose, A_transpose)
print(f"BᵀAᵀ = \n{BTA_transpose}")

are_transpose_equal = np.array_equal(AB_transpose, BTA_transpose)
print(f"\nIs (AB)ᵀ equal to BᵀAᵀ? {are_transpose_equal}")

if are_transpose_equal:
    print("✓ Proved: (AB)ᵀ = BᵀAᵀ")

print("\n" + "=" * 70)
print("Solving System of Linear Equations")
print("=" * 70)

# System of equations:
# 2x - 3y + z = -1
# x - y + 2z = -3  
# 3x + y - z = 9

print("System of equations:")
print("2x - 3y + z = -1")
print("x - y + 2z = -3")
print("3x + y - z = 9")

# Represent in matrix form: AX = B
A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

B_vector = np.array([-1, -3, 9])

print(f"\nMatrix A = \n{A_matrix}")
print(f"Vector B = {B_vector}")

# Method 1: Using inverse method (X = A⁻¹B)
print("\nMethod 1: Using Inverse Method (X = A⁻¹B)")
print("-" * 50)

try:
    # Calculate inverse of A
    A_inv = np.linalg.inv(A_matrix)
    print(f"A⁻¹ = \n{A_inv}")
    
    # Solve for X
    X_solution = np.dot(A_inv, B_vector)
    print(f"\nSolution X = A⁻¹B = {X_solution}")
    
    # Extract variables
    x, y, z = X_solution
    print(f"\nSolution:")
    print(f"x = {x:.2f}")
    print(f"y = {y:.2f}") 
    print(f"z = {z:.2f}")
    
    # Verify the solution
    print(f"\nVerification:")
    eq1 = 2*x - 3*y + z
    eq2 = x - y + 2*z
    eq3 = 3*x + y - z
    print(f"Equation 1: 2({x:.2f}) - 3({y:.2f}) + ({z:.2f}) = {eq1:.2f} (Expected: -1)")
    print(f"Equation 2: ({x:.2f}) - ({y:.2f}) + 2({z:.2f}) = {eq2:.2f} (Expected: -3)")
    print(f"Equation 3: 3({x:.2f}) + ({y:.2f}) - ({z:.2f}) = {eq3:.2f} (Expected: 9)")
    
except np.linalg.LinAlgError:
    print("Matrix A is singular, cannot use inverse method")

# Method 2: Using np.linalg.solve (recommended method)
print("\n" + "=" * 50)
print("Method 2: Using np.linalg.solve (Recommended)")
print("-" * 50)

try:
    X_solution_solve = np.linalg.solve(A_matrix, B_vector)
    print(f"Solution using np.linalg.solve = {X_solution_solve}")
    x_s, y_s, z_s = X_solution_solve
    print(f"x = {x_s:.2f}, y = {y_s:.2f}, z = {z_s:.2f}")
    
except np.linalg.LinAlgError:
    print("Matrix is singular, cannot solve")

print("\n" + "=" * 70)
print("Exploring np.linalg functions")
print("=" * 70)

# Additional linalg functions demonstration
print("\nAdditional np.linalg functions:")

# Determinant
det_A = np.linalg.det(A_matrix)
print(f"Determinant of A: {det_A:.2f}")

# Rank
rank_A = np.linalg.matrix_rank(A_matrix)
print(f"Rank of A: {rank_A}")

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A_matrix)
print(f"Eigenvalues of A: {eigenvalues}")
print(f"Eigenvectors of A: \n{eigenvectors}")

# Matrix norm
norm_A = np.linalg.norm(A_matrix)
print(f"Frobenius norm of A: {norm_A:.2f}")

# Condition number
cond_A = np.linalg.cond(A_matrix)
print(f"Condition number of A: {cond_A:.2f}")

print("\n" + "=" * 70)
print("Verification of Matrix Properties")
print("=" * 70)

# Additional verification
print("\nAdditional matrix property verification:")

# Verify (A⁻¹)⁻¹ = A
A_inv_inv = np.linalg.inv(A_inv)
print(f"(A⁻¹)⁻¹ = A? {np.allclose(A_inv_inv, A_matrix)}")

# Verify (Aᵀ)ᵀ = A
A_T_T = A_matrix.T.T
print(f"(Aᵀ)ᵀ = A? {np.array_equal(A_T_T, A_matrix)}")

# Verify determinant properties
det_A_inv = np.linalg.det(A_inv)
print(f"det(A⁻¹) = 1/det(A)? {np.allclose(det_A_inv, 1/det_A)}")

# Solve using least squares (for overdetermined systems)
print("\nLeast squares solution (alternative method):")
X_lstsq = np.linalg.lstsq(A_matrix, B_vector, rcond=None)[0]
print(f"Least squares solution: {X_lstsq}")
```
