```python
#8 Problem based on Popular Algorithm.
#8.1 Recursion:
#Task1
def sum_nested_list(nested_list):
  total = 0
  for element in nested_list:
    if isinstance(element, list):
      total += sum_nested_list(element) # Recursively sum the nested list
    else:
      total += element # Add the number to the total
  return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Task1", "\n")
print("Total sum:", result, "\n")
```

```python
#Task 2
#Generate All Permutations of a String:
def generate_permutations(s):
    res = []
    chars = list(s)
    used = [False] * len(chars)

    def backtrack(path):
        # Base case: when the permutation is complete
        if len(path) == len(chars):
            res.append(''.join(path))
            return

        seen = set()  # Avoid duplicate permutations if characters repeat
        for i in range(len(chars)):
            if used[i] or chars[i] in seen:
                continue

            seen.add(chars[i])        # Track characters used at this level
            used[i] = True            # Mark as used
            path.append(chars[i])     # Choose

            backtrack(path)           # Explore

            # Undo choice (backtrack)
            path.pop()
            used[i] = False

    backtrack([])
    return res
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
#Task 3 - Directory Size Calculation:
def calculate_directory_size(directory):
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
            # It's a subdirectory - recursively calculate its size
            total_size += calculate_directory_size(value)
        else:
            # It's a file - add its size directly
            total_size += value

    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test the function
total_size = calculate_directory_size(directory_structure)
print("Task3", "\n")
print(f"Total directory size: {total_size} KB", "\n")
```

```python
#8.2 Dynamic Programming:
#4. Example: Fibonacci Sequence:
#Solving Fibonacci Sequnce with Memoization {Top - Down Approach}.
def fibonacci(n, memo={}):
  """
  Returns:
  int: The nth Fibonacci number.
  Time Complexity:
  - The time complexity is O(n) because each Fibonacci number is computed only once.
  Space Complexity:
  - The space complexity is O(n) due to the memoization dictionary storing the results.
  """
  if n in memo:
    return memo[n]
  if n <= 1:
    return n
  memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
  return memo[n]
print(fibonacci(5))
print(fibonacci(10))
```

```python
# Bottom-up approach
#Build from smallest to largest:
def fibonacci(n):
  if n <= 1:
    return n
  dp = [0] * (n + 1)
  dp[1] = 1
  for i in range(2, n + 1):
    dp[i] = dp[i - 1] + dp[i - 2]
  return dp[n]
print(fibonacci(5))
print(fibonacci(10))
```

```python
#Task 1 - Coin Change Problem:
def min_coins(coins, amount):
  dp = [float('inf')] * (amount + 1)
  dp[0] = 0

  for coin in coins:
    for i in range(coin, amount + 1):
      dp[i] = min(dp[i], dp[i - coin] + 1)

  return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1, 2, 5], 11))
print(min_coins([2], 3))
```

```python
#Task 2 - Longest Common Subsequence (LCS):
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

print("Task 2 - Longest Common Subsequence:")
print("LCS length:", longest_common_subsequence("abcde", "ace"), "\n")
```

```python
#Task 3 - 0/1 Knapsack Problem:
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print("Task 3 - Knapsack Problem:")
print("Maximum value:", knapsack(weights, values, capacity))
```
