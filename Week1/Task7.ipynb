```python
#Task1
# Daily time (in hours): [study, entertainment, sleep]
# Study time classification
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
low_study = [] #3< hrs
moderate_study = [] #3-5 hrs
high_study = [] #5> hrs

# Classification loop for study
for study, entertainment, sleep in time_data:
    if study < 3:
        low_study.append(study)
    elif 3 <= study <= 5:
        moderate_study.append(study)
    else:
        high_study.append(study)
print("Low study hours:", low_study)
print("Moderate study hours:", moderate_study)
print("High study hours:", high_study)

#Task2
#Count the number of items in the low study
num_low_study = len(low_study)
print("Number of low study hours:", num_low_study)
#2
num_moderate_study = len(moderate_study)
print("Number of moderate study hours:", num_moderate_study)
#3
num_high_study = len(high_study)
print("Number of high study hours:", num_high_study)

#Task3
#Convert Study Hours to Minutes:
study_minutes = [study * 60 for study, _, _ in time_data]
print("total number of studies: ", len(study_minutes))
print("Study hours in minutes:", study_minutes)
```
